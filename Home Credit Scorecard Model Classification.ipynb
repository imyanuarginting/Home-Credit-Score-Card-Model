### **Missing Values**

```python
pd.set_option('display.max_rows', None)
df_isna = pd.DataFrame(df.isna().sum(), columns=['Missing Values'])
df_isna.reset_index(inplace=True)
df_isna.rename(columns={'index': 'Column Names'}, inplace=True)
df_isna['Percentage of Missing Values'] = (df_isna['Missing Values'] / df.shape[0]) * 100
df_isna.sort_values(by='Percentage of Missing Values', ascending=False, inplace=True)
df_isna.reset_index(drop=True, inplace=True)
df_isna = df_isna[df_isna['Percentage of Missing Values'] >= 47]
df_isna.style.background_gradient(cmap='Blues')
```

```python
pd.reset_option('display.max_rows')
```

```python
column_missing_values = df_isna['Column Names'].values
df.drop(df[column_missing_values], axis=1, inplace=True)
```

## **Modeling**

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_scaled, y_train)
y_pred_dt = dt.predict(X_test_scaled)

accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_score_dt = f1_score(y_test, y_pred_dt)
roc_auc_score_dt = roc_auc_score(y_test, y_pred_dt)

print("Decision Tree Results:")
print(f"Accuracy: {accuracy_dt:.3f}")
print(f"Precision: {precision_dt:.3f}")
print(f"Recall: {recall_dt:.3f}")
print(f"F1-Score: {f1_score_dt:.3f}")
print(f"ROC AUC Score: {roc_auc_score_dt:.3f}")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_scaled, y_train)
y_pred_rf = rf.predict(X_test_scaled)

accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_score_rf = f1_score(y_test, y_pred_rf)
roc_auc_score_rf = roc_auc_score(y_test, y_pred_rf)

print("Random Forest Results:")
print(f"Accuracy: {accuracy_rf:.3f}")
print(f"Precision: {precision_rf:.3f}")
print(f"Recall: {recall_rf:.3f}")
print(f"F1-Score: {f1_score_rf:.3f}")
print(f"ROC AUC Score: {roc_auc_score_rf:.3f}")
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=400, random_state=42)
lr.fit(X_train_scaled, y_train)
y_pred_lr = lr.predict(X_test_scaled)

accuracy_lr = accuracy_score(y_test, y_pred_lr)
precision_lr = precision_score(y_test, y_pred_lr)
recall_lr = recall_score(y_test, y_pred_lr)
f1_score_lr = f1_score(y_test, y_pred_lr)
roc_auc_score_lr = roc_auc_score(y_test, y_pred_lr)

print("Logistic Regression Results:")
print(f"Accuracy: {accuracy_lr:.3f}")
print(f"Precision: {precision_lr:.3f}")
print(f"Recall: {recall_lr:.3f}")
print(f"F1-Score: {f1_score_lr:.3f}")
print(f"ROC AUC Score: {roc_auc_score_lr:.3f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)

accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn)
recall_knn = recall_score(y_test, y_pred_knn)
f1_score_knn = f1_score(y_test, y_pred_knn)
roc_auc_score_knn = roc_auc_score(y_test, y_pred_knn)

print("K-Nearest Neighbors Results:")
print(f"Accuracy: {accuracy_knn:.3f}")
print(f"Precision: {precision_knn:.3f}")
print(f"Recall: {recall_knn:.3f}")
print(f"F1-Score: {f1_score_knn:.3f}")
print(f"ROC AUC Score: {roc_auc_score_knn:.3f}")
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train_scaled, y_train)
y_pred_nb = nb.predict(X_test_scaled)

accuracy_nb = accuracy_score(y_test, y_pred_nb)
precision_nb = precision_score(y_test, y_pred_nb)
recall_nb = recall_score(y_test, y_pred_nb)
f1_score_nb = f1_score(y_test, y_pred_nb)
roc_auc_score_nb = roc_auc_score(y_test, y_pred_nb)

print("Naive Bayes Results:")
print(f"Accuracy: {accuracy_nb:.3f}")
print(f"Precision: {precision_nb:.3f}")
print(f"Recall: {recall_nb:.3f}")
print(f"F1-Score: {f1_score_nb:.3f}")
print(f"ROC AUC Score: {roc_auc_score_nb:.3f}")
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(max_iter=400, random_state=42)
mlp.fit(X_train_scaled, y_train)
y_pred_mlp = mlp.predict(X_test_scaled)

accuracy_mlp = accuracy_score(y_test, y_pred_mlp)
precision_mlp = precision_score(y_test, y_pred_mlp)
recall_mlp = recall_score(y_test, y_pred_mlp)
f1_score_mlp = f1_score(y_test, y_pred_mlp)
roc_auc_score_mlp = roc_auc_score(y_test, y_pred_mlp)

print("Multi Layer Perceptron Results:")
print(f"Accuracy: {accuracy_mlp:.3f}")
print(f"Precision: {precision_mlp:.3f}")
print(f"Recall: {recall_mlp:.3f}")
print(f"F1-Score: {f1_score_mlp:.3f}")
print(f"ROC AUC Score: {roc_auc_score_mlp:.3f}")
```
